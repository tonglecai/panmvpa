## Step 2: Run pfm-mefmri

**Precision Functional Mapping - Multi-Echo fMRI** \
https://github.com/cjl2007/pfm-mefmri/blob/main/README.md#pfm-strategies \
This is the bash-wrapped python code created by Lynch, same pipeline as the PAN Dataset. \
It takes in unprocessed resting-state/anat data; a config file that includes several analysis preference, as well as a list of reference networks. \
It outputs consensus network label files (.dlabel.nii), confidence maps (.dscalar.nii), and community assignment tables (.csv)

to run: `bash bin/mefmri_pipeline.sh <SubjectDir> [ConfigFile]`

***However***, we only have access to *preprocessed* data, not raw data. So things will be a little more complicated. We need to run a subset of the modules instead of the entire thing. We can probably use the `START_FROM_MODULE` option in the config file to acheive this. I haven't look into too much detail but maybe we can start with the `vol2surf` module.
